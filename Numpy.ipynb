### Creating arrays

```python
a = np.array([1,2,3,4,5])
print(a)
```

```python
b = np.array([[1,2,3,4],[5,6,7,8]])
print(b)
```

```python
a = np.array([0,0,0,0])
print(a)
```

```python
zero = np.zeros((2,3))
print(zero)
```

```python
print(np.ones((2,3)))
```

```python
print(np.full((2,3),7))
```

```python
print(np.arange(0,10,2))
```

```python
print(np.eye(3)) #identity matrix
```

```python
rng = np.random.default_rng(seed=42)
print(rng.random((2,3)))
```

```python
print(rng.integers(1,10,size=(3,3))) #uniform
```

```python
print(rng.normal(loc=0, scale = 1,size = 5))
```

### Indexing and Slicing

```python
x = np.arange(10,20)
x
```

```python
print(x[0])
```

```python
print(x[5])
```

```python
m = np.arange(1,13).reshape(3,4)
m
```

```python
print(m[1,2])
```

```python
print(m[1])
```

```python
print(m[:,1])
```
