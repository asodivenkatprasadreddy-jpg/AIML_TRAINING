```python
def stats(nums):
    return min(nums), max(nums), sum(nums)/len(nums)
```

```python
low, high, avg = stats([10,20,30,40])
```

```python
print(low,high,avg)
```

```python
age = 15
if age >= 18:
    print('Adult')
else:
    print('minor')
```

```python
marks = 75
if marks >= 90:
    print("A")
elif marks >= 80 and marks < 90:
    print("B")
elif marks >= 70 and marks < 80:
    print("C")
else:
    print("D")
```

### List Comprehnsion

```python
numbers = [1,2,3,4,5]
squares = []

for x in numbers:
    squares.append(x)
print(squares)
```

```python
squares = [x**2 for x in range(5)]
print(squares)
```

```python
even = [x for x in range(20) if x % 2 == 0]
print(even)
```

```python
words = ['python','machinelearning','datascience']
upper = [w.upper() for w in words]
print(upper)
```
